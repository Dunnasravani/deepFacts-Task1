```python
import json
import mysql.connector
import requests
from tabulate import tabulate
from flask import Flask, jsonify

app = Flask(__name__)

# Function to fetch data from the Geonames API and insert it into the MySQL database
def fetch_and_store_data():
    # Step 1: Fetch the JSON data from the API
    url = "http://api.geonames.org/searchJSON?country=IN&featureClass=P&maxRows=100&username=srirampalika23"
    response = requests.get(url)
    data = response.json()

    # Step 2: Connect to the MySQL database on port 3307
    conn = mysql.connector.connect(
        host='localhost',
        user='root',    # Replace with your MySQL username
        password='root', # Replace with your MySQL password
        database='geolocations', # Replace with your MySQL database name
        port=3307                 # Connect to port 3307
    )
    cursor = conn.cursor()

    # Step 3: Create the table if it doesn't exist
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS Geonames (
            name VARCHAR(255) NOT NULL,
            lat DOUBLE NOT NULL
        )
    ''')

    # Step 4: Insert data into the table
    for city in data['geonames']:
        name = city['name']
        lat = float(city['lat'])  # Convert lat to float for MySQL
        sql = "INSERT INTO Geonames (name, lat) VALUES (%s, %s)"
        cursor.execute(sql, (name, lat))
    
    # Commit the transaction
    conn.commit()

    # Step 5: Close the connection
    cursor.close()
    conn.close()

@app.route('/geonames', methods=['GET'])
def get_geonames():
    # Connect to the MySQL database
    conn = mysql.connector.connect(
        host='localhost',
        user='root',    # Replace with your MySQL username
        password='root', # Replace with your MySQL password
        database='geolocations', # Replace with your MySQL database name
        port=3307                 # Connect to port 3307
    )
    cursor = conn.cursor()

    # Step 6: Query the data from the Geonames table
    cursor.execute('SELECT name, lat FROM Geonames')

    # Fetch all rows from the result
    rows = cursor.fetchall()

    # Step 7: Prepare the data for JSON response
    data = [{"name": row[0], "latitude": row[1]} for row in rows]

    # Close the connection
    cursor.close()
    conn.close()

    # Return the data as a JSON response
    return jsonify(data)

if __name__ == '__main__':
    fetch_and_store_data()  # Fetch and store data from the API on startup
    app.run(debug=True, port=5000)  # Runs the app on port 5000
```
